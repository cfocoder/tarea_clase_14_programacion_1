```python
import sqlite3
import pandas as pd
from pathlib import Path

# Conexión a la base de datos SQLite
db_path = Path("DB_Propia.db")
conexion = sqlite3.connect(str(db_path))

print("✓ Conexión a DB_Propia.db exitosa")
```

## 1. SELECT - Obtener todos los registros

```python
query_select_empleados = "SELECT * FROM empleados"
df_empleados = pd.read_sql(query_select_empleados, conexion)
df_empleados
```

```python
query_select_proyectos = "SELECT * FROM proyectos"
df_proyectos = pd.read_sql(query_select_proyectos, conexion)
df_proyectos
```

```python
query_select_clientes = "SELECT * FROM clientes"
df_clientes = pd.read_sql(query_select_clientes, conexion)
df_clientes
```

## 2. WHERE - Filtrar registros con condiciones

```python
# Empleados con salario mayor a 50000
query_where_salario = "SELECT nombre, departamento, salario FROM empleados WHERE salario > 50000"
df_salario_alto = pd.read_sql(query_where_salario, conexion)
print("Empleados con salario > 50000:")
df_salario_alto
```

```python
# Clientes de una ciudad específica
query_where_ciudad = "SELECT nombre_cliente, ciudad, email FROM clientes WHERE ciudad = 'Madrid'"
df_clientes_madrid = pd.read_sql(query_where_ciudad, conexion)
print("Clientes en Madrid:")
df_clientes_madrid
```

## 3. ORDER BY - Ordenar resultados

```python
# Empleados ordenados por salario descendente
query_order_by_salario = "SELECT nombre, departamento, salario FROM empleados ORDER BY salario DESC"
df_empleados_ordenados = pd.read_sql(query_order_by_salario, conexion)
print("Empleados ordenados por salario (Mayor a Menor):")
df_empleados_ordenados
```

```python
# Proyectos ordenados por presupuesto ascendente
query_order_by_presupuesto = "SELECT nombre_proyecto, presupuesto, duracion_meses FROM proyectos ORDER BY presupuesto ASC"
df_proyectos_ordenados = pd.read_sql(query_order_by_presupuesto, conexion)
print("Proyectos ordenados por presupuesto (Menor a Mayor):")
df_proyectos_ordenados
```

## 4. AGGREGATE FUNCTIONS - COUNT, SUM, AVG, MIN, MAX

```python
# COUNT - Contar el total de empleados
query_count = "SELECT COUNT(*) as total_empleados FROM empleados"
df_count = pd.read_sql(query_count, conexion)
print("Total de empleados:")
df_count
```

```python
# SUM - Suma total de presupuestos
query_sum = "SELECT SUM(presupuesto) as presupuesto_total FROM proyectos"
df_sum = pd.read_sql(query_sum, conexion)
print("Presupuesto total de todos los proyectos:")
df_sum
```

```python
# AVG - Promedio de salarios
query_avg = "SELECT AVG(salario) as salario_promedio FROM empleados"
df_avg = pd.read_sql(query_avg, conexion)
print("Salario promedio:")
df_avg
```

```python
# MIN - Salario mínimo
query_min = "SELECT MIN(salario) as salario_minimo FROM empleados"
df_min = pd.read_sql(query_min, conexion)
print("Salario mínimo:")
df_min
```

```python
# MAX - Salario máximo
query_max = "SELECT MAX(salario) as salario_maximo FROM empleados"
df_max = pd.read_sql(query_max, conexion)
print("Salario máximo:")
df_max
```

## 5. GROUP BY - Agrupar datos

```python
# GROUP BY - Agrupar empleados por departamento y calcular promedio de salario
query_group_by = """
SELECT departamento, COUNT(*) as cantidad_empleados, AVG(salario) as salario_promedio 
FROM empleados 
GROUP BY departamento
"""
df_group_by = pd.read_sql(query_group_by, conexion)
print("Empleados agrupados por departamento:")
df_group_by
```

```python
# GROUP BY - Proyectos agrupados por estado
query_group_by_estado = """
SELECT estado, COUNT(*) as cantidad_proyectos, SUM(presupuesto) as presupuesto_total 
FROM proyectos 
GROUP BY estado
"""
df_group_by_estado = pd.read_sql(query_group_by_estado, conexion)
print("Proyectos agrupados por estado:")
df_group_by_estado
```

## 6. HAVING - Filtrar grupos

```python
# HAVING - Departamentos con salario promedio mayor a 55000
query_having = """
SELECT departamento, COUNT(*) as cantidad_empleados, AVG(salario) as salario_promedio 
FROM empleados 
GROUP BY departamento
HAVING AVG(salario) > 55000
"""
df_having = pd.read_sql(query_having, conexion)
print("Departamentos con salario promedio > 55000:")
df_having
```

## 7. Análisis de la tabla Maratón NY - SELECT

```python
# Obtener primeros 10 corredores de la tabla maratón_ny
query_maraton_select = "SELECT * FROM maraton_ny LIMIT 10"
df_maraton_sample = pd.read_sql(query_maraton_select, conexion)
print("Primeros 10 corredores de la Maratón NY:")
df_maraton_sample
```

## 8. Maratón NY - WHERE (Filtrar por genero y tiempo)

```python
# Corredores de género femenino
query_maraton_where_genero = "SELECT corredor, lugar, genero, edad, pais_origen, tiempo_minutos FROM maraton_ny WHERE genero = 'F' LIMIT 10"
df_maraton_mujeres = pd.read_sql(query_maraton_where_genero, conexion)
print("Corredoras de género femenino (primeras 10):")
df_maraton_mujeres
```

```python
# Corredores con tiempo menor a 300 minutos
query_maraton_tiempo = "SELECT corredor, lugar, genero, edad, pais_origen, tiempo_minutos FROM maraton_ny WHERE tiempo_minutos < 300 LIMIT 10"
df_maraton_rapidos = pd.read_sql(query_maraton_tiempo, conexion)
print("Corredores con tiempo < 300 minutos (primeros 10):")
df_maraton_rapidos
```

## 9. Maratón NY - ORDER BY (Ordenar corredores)

```python
# Top 10 corredores más rápidos (menor tiempo)
query_maraton_order_tiempo = "SELECT corredor, lugar, genero, edad, tiempo_minutos FROM maraton_ny ORDER BY tiempo_minutos ASC LIMIT 10"
df_top_rapidos = pd.read_sql(query_maraton_order_tiempo, conexion)
print("Top 10 corredores más rápidos:")
df_top_rapidos
```

```python
# Top 10 corredores más lentos (mayor tiempo)
query_maraton_order_lento = "SELECT corredor, lugar, genero, edad, tiempo_minutos FROM maraton_ny ORDER BY tiempo_minutos DESC LIMIT 10"
df_top_lentos = pd.read_sql(query_maraton_order_lento, conexion)
print("Top 10 corredores más lentos:")
df_top_lentos
```

## 10. Maratón NY - AGGREGATE FUNCTIONS

```python
# Total de corredores en la maratón
query_maraton_count = "SELECT COUNT(*) as total_corredores FROM maraton_ny"
df_maraton_count = pd.read_sql(query_maraton_count, conexion)
print("Total de corredores:")
df_maraton_count
```

```python
# Tiempo promedio de la maratón
query_maraton_avg = "SELECT AVG(tiempo_minutos) as tiempo_promedio FROM maraton_ny"
df_maraton_avg = pd.read_sql(query_maraton_avg, conexion)
print("Tiempo promedio (en minutos):")
df_maraton_avg
```

```python
# Tiempo mínimo y máximo
query_maraton_minmax = "SELECT MIN(tiempo_minutos) as tiempo_minimo, MAX(tiempo_minutos) as tiempo_maximo FROM maraton_ny"
df_maraton_minmax = pd.read_sql(query_maraton_minmax, conexion)
print("Tiempo mínimo y máximo (en minutos):")
df_maraton_minmax
```

## 11. Maratón NY - GROUP BY (Agrupar por género y país)

```python
# Agrupar por género y calcular estadísticas
query_maraton_group_genero = """
SELECT genero, 
       COUNT(*) as cantidad_corredores,
       AVG(edad) as edad_promedio,
       AVG(tiempo_minutos) as tiempo_promedio
FROM maraton_ny 
GROUP BY genero
"""
df_maraton_por_genero = pd.read_sql(query_maraton_group_genero, conexion)
print("Corredores agrupados por género:")
df_maraton_por_genero
```

```python
# Agrupar por país de origen
query_maraton_group_pais = """
SELECT pais_origen, 
       COUNT(*) as cantidad_corredores,
       AVG(tiempo_minutos) as tiempo_promedio
FROM maraton_ny 
GROUP BY pais_origen
ORDER BY cantidad_corredores DESC
LIMIT 10
"""
df_maraton_por_pais = pd.read_sql(query_maraton_group_pais, conexion)
print("Top 10 países con más corredores:")
df_maraton_por_pais
```

## 12. Maratón NY - HAVING (Filtrar grupos)

```python
# Países con tiempo promedio menor a 350 minutos
query_maraton_having = """
SELECT pais_origen, 
       COUNT(*) as cantidad_corredores,
       AVG(tiempo_minutos) as tiempo_promedio
FROM maraton_ny 
GROUP BY pais_origen
HAVING AVG(tiempo_minutos) < 350
ORDER BY tiempo_promedio ASC
"""
df_maraton_having = pd.read_sql(query_maraton_having, conexion)
print("Países con tiempo promedio < 350 minutos:")
df_maraton_having
```

## 14. Drivers para Conectarse a Diferentes Bases de Datos

### Drivers disponibles para diferentes bases de datos:

**Para MySQL:**
- **mysql-connector-python**: Driver oficial de MySQL
- **PyMySQL**: Driver puro de Python (sin dependencias externas)
- **mysqlclient** (MySQLdb): Wrapper de libmysqlclient (más rápido pero requiere librerías del sistema)

**Para PostgreSQL:**
- **psycopg2**: Driver más popular y recomendado
- **psycopg3**: Versión más nueva y moderna

**Para SQL Server:**
- **pyodbc**: Driver genérico ODBC (también funciona con MySQL)
- **pymssql**: Driver específico para SQL Server

**Para SQLite (usado en esta tarea):**
- **sqlite3**: Integrado en Python por defecto

**Para Oracle:**
- **cx_Oracle**: Driver oficial de Oracle

**Para otras bases de datos:**
- **psycopg2**: PostgreSQL
- **pymongo**: MongoDB
- **cassandra-driver**: Apache Cassandra
